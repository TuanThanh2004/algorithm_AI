```python
import string

remove_charaters = '\t""'+ string.punctuation
```

```python
remove_charaters
```

```python
def text_normaliza(text):
    text = text.lower()
    text = text.strip()
    text = text.replace('\n', ' ')
    for char in remove_charaters:
        text = text.replace(char, ' ')

    return text
```

```python
def create_dictionary(corous):
    dictionary = []
    # for paragraph in corous:
    #     paragraph = text_normaliza(paragraph)
    #     tokens = paragraph.split()
    #     for token in tokens:
    #         if token not in dictionary:
    #             dictionary.append(token)
    paragraph = text_normaliza(corous)
    tokens = paragraph.split()
    for token in tokens:
        if token not in dictionary:
            dictionary.append(token)

    return dictionary
```

```python
def vectorize(
    text,
    dictionary,
    unknow_token_id
):
    text = text_normaliza(text)
    tokens = text.split()
    vector = [
        dictionary.index(token) if token in dictionary else unknow_token_id for token in tokens
    ]

    return vector
```
